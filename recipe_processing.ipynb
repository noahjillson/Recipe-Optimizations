```python
import pandas as pd
```

```python
recipes = pd.read_json("recipes_raw_nosource_fn.json").dropna(axis=1)
```

```python
recipes.sort_values(by=['ingredients'], axis=1, ascending=False, key=lambda x: [len(y) for y in x])
```

```python
i = 0
recipes_lst = []
rdict = {}
for key in recipes.keys():
    if i > 100:
        break
    i+=1
    recipes_lst.append(recipes[key])
    rdict[key] = recipes[key]
```

```python
df = pd.concat(recipes_lst)
```

```python
df.to_json('top100.json', orient='records')
```
